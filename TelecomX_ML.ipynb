```python
import seaborn as sns
import matplotlib.pyplot as plt

col_categoricas = datos.select_dtypes(include='object').columns

n_rows = len(col_categoricas)

plt.figure(figsize=(12, n_rows*4))

for i, columna in enumerate(col_categoricas):

    avg_churn = datos.groupby(columna)['baja_servicio'].mean().reset_index()

    # gráfico churn
    plt.subplot(n_rows, 2, i*2 + 1)
    sns.barplot(x=columna, y='baja_servicio', data=avg_churn)
    plt.title(f'{columna} vs Churn')
    plt.xticks(rotation=45)

    # gráfico cantidad
    plt.subplot(n_rows, 2, i*2 + 2)
    sns.countplot(data=datos, x=columna)
    plt.title(f'Número de {columna}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

El género no influye en la cancelación.

Tener múltiples líneas no cambia mucho el churn.

Los clientes con fibra óptica cancelan más que los de otros tipos de internet.

Los clientes sin servicios adicionales (seguridad, respaldo, soporte) cancelan más.

Los servicios de streaming casi no influyen en el churn.

El contrato mes a mes tiene la mayor cancelación, mientras que los contratos de 1 o 2 años reducen el churn.

El electronic check tiene la mayor tasa de cancelación entre los métodos de pago.

```python
ax = sns.countplot(x='baja_servicio', data=datos)

total = len(datos)

for p in ax.patches:
    porcentaje = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(porcentaje, (x, y), ha='center', va='bottom')

plt.title('Distribución de cancelaciones')
plt.xlabel('Baja de servicio')
plt.ylabel('Cantidad')
plt.show()
```

###3.Modelado Predictivo

####3.1 Separación de Datos

```python
#Generando un modelo base de comparación con 30% de los datos de prueba

from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split

X = datos.drop(['baja_servicio'],axis=1)
y= datos['baja_servicio']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=5)

baseline = DummyClassifier()
baseline.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = baseline.predict(X_test)

accuracy_score(y_test,y_pred)
```

```python
baseline.score(X_test,y_test)
```

####3.2 Creación de Modelos

#####Árbol de decisión

```python
from sklearn.tree import DecisionTreeClassifier

modelo_arbol = DecisionTreeClassifier(max_depth=10)
modelo_arbol.fit(X_train,y_train)

```

```python
modelo_arbol.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = modelo_arbol.predict(X_test)
print(classification_report(y_test, y_pred))
```

Detecta bien a los clientes que no cancelan (83%)

Detecta solo 55% de los que cancelan

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel("Predicción")
plt.ylabel("Valor real")
plt.title("Matriz de confusión")

plt.show()
```

```python
#mejorando modelo disminuyendo error tipo II

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=5)

X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

modelo = DecisionTreeClassifier(max_depth=10, random_state=5)

modelo.fit(X_train_res, y_train_res)

y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel("Predicción")
plt.ylabel("Valor real")
plt.title("Matriz de confusión")

plt.show()
```

#####Logistic Regression

```python
#NORMALIZANDO

from sklearn.preprocessing import MinMaxScaler

normalizacion = MinMaxScaler()
X_train_normalizado = normalizacion.fit_transform(X_train)
X_test_normalizado = normalizacion.transform(X_test)

X_train_normalizado = pd.DataFrame(
    X_train_normalizado,
    columns=X_train.columns
)

X_test_normalizado = pd.DataFrame(
    X_test_normalizado,
    columns=X_test.columns
)
```

```python

from sklearn.linear_model import LogisticRegression


modelo_log = LogisticRegression(max_iter=1000)
modelo_log.fit(X_train_normalizado, y_train)


y_pred = modelo_log.predict(X_test_normalizado)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel("Predicción")
plt.ylabel("Valor real")
plt.title("Matriz de confusión")

plt.show()
```

Se evaluaron diferentes modelos para predecir la cancelación de clientes utilizando métricas como accuracy, precision, recall, F1-score y la matriz de confusión.

El modelo de árbol de decisión combinado con SMOTE mostró una mejor capacidad para detectar clientes que cancelan el servicio, reduciendo el número de falsos negativos.

Aunque el modelo de Logistic Regression presentó menos falsos positivos, su capacidad para identificar churn fue menor.

Por esta razón, el modelo de árbol de decisión con balanceo de datos se considera el modelo con mejor desempeño para este problema.

```python
# Analizar coeficientes del modelo de Regresión Logística

coeficientes = pd.DataFrame({
    "Variable": X_train_normalizado.columns,
    "Coeficiente": modelo_log.coef_[0]
})

coeficientes = coeficientes.sort_values(by="Coeficiente", ascending=False)

print("Variables que aumentan la probabilidad de cancelación:")
display(coeficientes.head(10))

print("\nVariables que reducen la probabilidad de cancelación:")
display(coeficientes.tail(10))
```

El análisis de importancia de variables en Random Forest muestra que el tipo de contrato, el tiempo de permanencia del cliente y los cargos asociados al servicio son factores clave en la predicción de cancelación.

Por otro lado, en el modelo de Regresión Logística se observa que variables como el contrato mensual y el tipo de servicio de internet aumentan la probabilidad de churn, mientras que variables relacionadas con servicios adicionales como seguridad online o soporte técnico tienden a reducir la probabilidad de cancelación.

Estos resultados sugieren que los clientes con contratos más flexibles y menor permanencia presentan mayor riesgo de abandonar el servicio.

# Informe de análisis de cancelación de clientes (Churn)

## Objetivo del análisis

El objetivo de este análisis fue identificar los factores que influyen en la cancelación de clientes y evaluar distintos modelos de machine learning para predecir el churn. Para ello se realizaron análisis exploratorios de datos, transformación de variables categóricas mediante técnicas de codificación, entrenamiento de diferentes modelos y evaluación de su desempeño utilizando métricas como accuracy, precision, recall, F1-score y la matriz de confusión.

## Evaluación de modelos

Durante el desarrollo del análisis se implementaron diferentes modelos de clasificación para predecir la cancelación de clientes:

- Dummy Classifier (modelo base)
- Árbol de decisión
- Árbol de decisión con balanceo de datos (SMOTE)
- Regresión Logística

El modelo Dummy obtuvo una exactitud cercana al **73%**, lo cual refleja principalmente la proporción de clientes que no cancelan el servicio.

El modelo de **árbol de decisión** mejoró el rendimiento alcanzando aproximadamente **76% de accuracy**, lo que indica que el modelo logró identificar patrones relevantes en los datos.

Posteriormente, se aplicó la técnica **SMOTE** para balancear las clases del conjunto de entrenamiento. Esto permitió mejorar la detección de clientes que cancelan el servicio, reduciendo el número de falsos negativos de aproximadamente **264 a 225 casos**, y aumentando el número de cancelaciones correctamente detectadas de **324 a 363 clientes**. Esto representa una mejora en el **recall de churn de aproximadamente 55% a 62%**.

El modelo de **Regresión Logística** presentó una matriz de confusión con **302 clientes que cancelaron correctamente identificados y 286 falsos negativos**, mostrando menor capacidad para detectar churn en comparación con el árbol balanceado, aunque con menos falsos positivos.

Dado que en problemas de churn es especialmente importante identificar clientes que podrían abandonar el servicio, el modelo de **árbol de decisión con SMOTE mostró el mejor desempeño general**.

## Factores que influyen en la cancelación

A partir del análisis de importancia de variables obtenido mediante Random Forest y de los coeficientes del modelo de Regresión Logística, se identificaron varios factores relevantes que influyen en la cancelación de clientes.

Entre los principales factores se encuentran:

### Tipo de contrato
El tipo de contrato es uno de los factores más importantes. Los clientes con contrato **Month-to-month** presentan mayor probabilidad de cancelar el servicio debido a la flexibilidad para cambiar de proveedor.

### Permanencia del cliente
La variable **permanencia_mensual** muestra que los clientes con menor tiempo en la empresa tienen mayor probabilidad de cancelar. En cambio, los clientes con mayor permanencia tienden a mantenerse más tiempo en el servicio.

### Costos del servicio
Variables como **cargo_total, cargo_mensual y cargo_diario** también influyen en la cancelación, lo que sugiere que el costo del servicio puede afectar la decisión de continuar o abandonar la compañía.

### Tipo de servicio de internet
El tipo de conexión, especialmente **Fiber optic**, aparece como una variable relevante en la predicción del churn, lo que puede estar relacionado con expectativas más altas del servicio o mayor competencia en ese segmento del mercado.

### Servicios adicionales
Los clientes que cuentan con servicios adicionales como **seguridad online, soporte técnico o respaldo online** presentan menor probabilidad de cancelar el servicio, lo que indica un mayor nivel de fidelización.

## Estrategias de retención basadas en los resultados

A partir de los factores identificados, se pueden proponer diversas estrategias para reducir la cancelación de clientes:

### Incentivar contratos de mayor duración
Dado que los clientes con contratos mensuales presentan mayor tasa de cancelación, se podrían ofrecer **descuentos o beneficios** para incentivar la contratación de planes de mayor duración.

### Programas de fidelización para nuevos clientes
Los clientes con menor permanencia presentan mayor riesgo de cancelación. Implementar **programas de fidelización durante los primeros meses de servicio** podría ayudar a mejorar la retención.

### Mejora en la calidad del servicio
Debido a la influencia del tipo de servicio de internet en la cancelación, mejorar la **calidad y estabilidad del servicio** puede contribuir a reducir el churn.

### Promoción de servicios adicionales
Ofrecer paquetes que incluyan **seguridad online, soporte técnico o respaldo de datos** puede aumentar el valor percibido por el cliente y disminuir la probabilidad de cancelación.
